```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
error_data = pd.read_csv("computed_errors_112.csv")
error_data
```

```python
total_angular_error_list = error_data['Angular_Error_Yaw_Total']
total_angular_error_list_abs = total_angular_error_list.abs()
total_angular_error_list_abs
```

```python
print(f"Final average error over {len(total_angular_error_list_abs)} samples: {np.mean(total_angular_error_list_abs)}")
```

```python
fix, ax = plt.subplots()
ax.plot([i for i in range(len(total_angular_error_list_abs))], total_angular_error_list_abs)
plt.show()
```
